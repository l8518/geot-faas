```python
import os
import pandas as pd
```

```python
OUTPUT_FOLDER = "data"
INPUT_BATCHES_FOLDER = "./data/batches/"
INPUT_MSTATS_FOLDER = "./data/measurements/"
```

```python
batches = os.listdir(INPUT_BATCHES_FOLDER)
batches.sort()
mstats = os.listdir(INPUT_MSTATS_FOLDER)
mstats.sort()
batches = batches[:2]
mstats = mstats[:2]
batches, mstats
```

```python
def concat(folder, files):
    pds = []
    for f in files:
        fpath = os.path.join(folder, f)
        pds.append(pd.read_parquet(fpath))
    dataset = pd.concat(pds)
    return dataset

batchdf = concat(INPUT_BATCHES_FOLDER, batches)
mstatsdf = concat(INPUT_MSTATS_FOLDER, batches)
```

```python
batchdf = batchdf.sort_values(by=['driver_invocation', 'workload_invocation', 'provider', 'region', '1_run_id', '2_thread_id'])
batchdf['driver_invocation'] = batchdf['driver_invocation']+"00"
batchdf['workload_invocation'] = batchdf['workload_invocation']+"00"
batchdf.to_parquet(os.path.join(OUTPUT_FOLDER,"batchdf.parquet"), index=False)
mstatsdf.to_parquet(os.path.join(OUTPUT_FOLDER,"mstatsdf.parquet"), index=False)
batchdf, mstatsdf
```

```python
sample_batchdf = batchdf.groupby(['driver_invocation', 'provider', 'region']).sample(frac=0.1, random_state=10)
sample_batchdf.to_parquet(os.path.join(OUTPUT_FOLDER, "batchdf-0.1.parquet"), index=False)
sample_batchdf = batchdf.groupby(['driver_invocation', 'provider', 'region']).sample(frac=0.01, random_state=10)
sample_batchdf.to_parquet(os.path.join(OUTPUT_FOLDER, "batchdf-0.01.parquet"), index=False)
```
