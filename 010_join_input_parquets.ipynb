```python
import os
import pandas as pd
```

```python
INPUT_PARQUET_FOLDER = "./input_data/"
```

```python
input_parquets = os.listdir(INPUT_PARQUET_FOLDER)
input_parquets
```

```python
batch_pds = []
for ip in input_parquets:
    ip_path = os.path.join(INPUT_PARQUET_FOLDER, ip)
    batch_pds.append(pd.read_parquet(ip_path))
dataset = pd.concat(batch_pds)
dataset
```

```python
dataset = dataset.sort_values(by=['driver_invocation', 'workload_invocation', 'provider', 'region', '1_run_id', '2_thread_id'])
dataset['driver_invocation'] = dataset['driver_invocation']+"00"
dataset['workload_invocation'] = dataset['workload_invocation']+"00"
dataset.to_parquet("raw-dataset.parquet", index=False)
dataset
```

```python
sample_dataset = dataset.groupby(['driver_invocation', 'provider', 'region']).sample(frac=0.1, random_state=10)
sample_dataset.to_parquet("raw-dataset-sampled.parquet", index=False)
sample_dataset
```
